```python
import random

class Account:
    account_count = 0
# Q1: 생성자 구현
    def __init__(self, owner, balance=0):
        self.bank_name = "SC은행"  # Q2: 클래스 변수 선언
        self.owner = owner  # 예금주
        self.balance = balance  # 초기 잔액
        self.account_number = f"{random.randint(100, 999)}-{random.randint(10, 99)}-{random.randint(100000, 999999)}"  # 계좌번호 랜덤 생성
        Account.account_count += 1  # 계좌 생성 시마다 클래스 변수 account_count 증가
        self.deposit_count = 0  # 입금 횟수 초기화
        self._deposit_history = []  # 입금 내역 초기화
        self._withdraw_history = []  # 출금 내역 초기화
# Q3: 클래스 변수 출력 메서드
    @classmethod
    def get_account_num(cls):
        return cls.account_count
# Q4: 입금 메서드
    def deposit(self, amount):
        if amount < 1:
            print("입금은 최소 1원 이상 가능합니다.")
            return
        self.balance += amount
        self.deposit_count += 1  # 입금 횟수 증가
        self._deposit_history.append(amount)  # 입금 내역에 금액 추가
        self.apply_interest()
# Q5: 출금 메서드
    def withdraw(self, amount):
        if amount > self.balance:
            print("계좌잔고 이상으로 출금할 수 없습다.")
            return
        self.balance -= amount
        self._withdraw_history.append(amount)  # 출금 내역에 금액 추가
# Q6: 정보 출력 메서드
    def display_info(self):
        print(f"은행이름: {self.bank_name}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance:,.0f}원")

    def get_deposit_history(self):
        print("입금 내역:", self._deposit_history)

    def get_withdraw_history(self):
        print("출금 내역:", self._withdraw_history)
# Q7: 5회 입금 시 이자 지급
    def apply_interest(self):
        # 입금 횟수가 5회가 될 때 이자 지급
        if self.deposit_count >= 5:
            interest = self.balance * 0.01  # 1% 이자
            self.balance += interest  # 이자 추가
            print(f"이자 지급: {interest:,.0f}원")
            self.deposit_count = 0  # 입금 횟수 초기화

# Q8: 여러 객체 생성
account1 = Account("차정은", 1000)
account2 = Account("박광석", 30000000)
account3 = Account("조웅제", 50000000)

# 여러 객체 생성 후 리스트에 저장
accounts = [account1, account2, account3]

# 계좌 개수 출력
print("생성된 계좌 수:", Account.get_account_num())

# 계좌 정보 출력
for account in accounts:
    account.display_info()

# 입금 및 출금 테스트
account1.deposit(50000)
account1.withdraw(200000)

# 입금과 출금 내역 출력
account1.get_deposit_history()
account1.get_withdraw_history()

# Q9: 잔고가 100만원 이상인 고객의 정보 출력
for account in accounts:
    if account.balance >= 1000000:
        account.display_info()

# Q10: 입금 및 출금 내역 테스트
accounts[0].deposit(100000)
accounts[0].deposit(200000)
accounts[0].withdraw(150000)
```

```python
import random

# Q1. Character 클래스 만들기
class Character:
    def __init__(self, name, level, health, attack, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        effective_damage = max(0, damage - self.defense)  # 방어력만큼 데미지 차감
        self.health -= effective_damage

    def attack_target(self, target):
        damage = random.randint(1, self.attack)  # 1부터 공격력 사이의 랜덤한 데미지
        target.take_damage(damage)
        return damage

# Q2. Player 클래스 만들기
class Player(Character):
    def __init__(self, name):
        super().__init__(name, 1, 100, 25, 5)
        self.experience = 0

    def gain_experience(self, amount):
        self.experience += amount

    def level_up(self):
        while self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.experience -= 50
            print(f"{self.name} has leveled up! Current level: {self.level}, Attack: {self.attack}, Defense: {self.defense}")

# Q2. Monster 클래스
class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack, defense)

# Q3. battle 함수 만들기
def battle(player, monster):
    print(f"\nA wild {monster.name} appeared! Level: {monster.level}")
    while player.is_alive() and monster.is_alive():
        # Player attacks Monster
        damage_to_monster = player.attack_target(monster)
        print(f"{player.name} 이 {monster.name} 에게 {damage_to_monster} 만큼 공격했다!")

        # Monster attacks Player
        if monster.is_alive():
            damage_to_player = monster.attack_target(player)
            print(f"{monster.name} 이 {player.name} 에게 {damage_to_player} 만큼 공격했다!")

    # 결과 출력
    if player.is_alive():
        print("전투 승리!")
        print(f"{monster.name}을 이겼다!")
        player.gain_experience(monster.level * 20)
        player.level_up()
    else:
        print("전투 패배..")

# Q4. main 함수 만들기
def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}

    # 사용자로부터 이름 입력받기
    player_name = input("플레이어 이름을 입력하세요: ")
    player = Player(player_name)

    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)

        battle(player, monster)

        # 플레이어가 사망하면 게임 종료
        if not player.is_alive():
            print("게임오버")
            break

if __name__ == "__main__":
    main()
```
